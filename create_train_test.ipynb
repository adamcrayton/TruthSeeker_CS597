```python
import pandas as pd
import numpy as np
df = pd.read_csv('./dataset/TruthSeeker2023/Truth_Seeker_Model_Dataset.csv')
```

```python
df = df[~df['5_label_majority_answer'].isin(['NO MAJORITY'])]
```

```python
num_samples = len(df)
```

```python
permuted_indices = np.random.permutation(np.arange(num_samples))
```

```python
num_train = int(num_samples * 0.8)
train_indices = permuted_indices[:num_train]
test_indices = permuted_indices[num_train:]
```

```python
len(train_indices), len(test_indices)
```

```python
train_df = pd.DataFrame({"index": train_indices})
test_df = pd.DataFrame({"index": test_indices})
```

```python
train_df.to_csv("dataset/TruthSeeker2023/train_indices.csv", index=False)
```

```python
test_df.to_csv("dataset/TruthSeeker2023/test_indices.csv", index=False)
```
